```python
import requests
response=requests.get(
    'https://data.nasa.gov/resource/gh4g-9sfh.json',params={'$limit':50_000}
)
if response.ok:
  payload=response.json()
else:
  print(f'Request was not succcessful and returned code:{response.status_code}.')
  payload=None
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()

x=np.linspace(0,5,10)
#print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set_title('Multiple Lines')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
plt.show()
```

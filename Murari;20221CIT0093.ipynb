```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')
ax.set_title('multiple lines')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.legend()
#ax.legend(loc=1)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')
ax.plot(x,np.tan(x),label='tan')
ax.set_title('multiple lines')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.legend()
#ax.legend(loc=1)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dashed')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashed')
ax.set_title('multiple lines')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.legend()
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
x=np.linespace(0,10,1000)
axs[0,0].plot(x,np,sin(x))
axs[0,1].plot(x,np,cos(x),color='maroon')
axs[1,0].plot(x,np,sin(x+2),color='blue')
axs[1,1].plots(x,np,sin(x+4),color='green');
```
